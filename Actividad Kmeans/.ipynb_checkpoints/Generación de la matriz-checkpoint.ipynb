```python
#Importar datos de los archivos
import pandas as pd
import numpy as np

#Cargar primer archivo
doc_1 = pd.read_csv('sims_All_Genomes_13V_0_536.txt', sep=',',header=None)
#Eliminar los ultimos 36 elementos del documento 1
doc_1 = doc_1.drop(range(500,536))

#Cargar segundo archivo
doc_2 = pd.read_csv('sims_All_Genomes_13V_500_800.txt', sep=',',header=None)
#Eliminar los ultimos 23 elementos del documento 2
doc_2 = doc_2.drop(range(300,323))

#Cargar tercer archivo
doc_3 = pd.read_csv('sims_All_Genomes_13V_800_1300.txt', sep=',',header=None)
#Eliminar los ultimos 9 elementos del documento 3
doc_3 = doc_3.drop(range(500,509))

#Cargar cuarto archivo
doc_4 = pd.read_csv('sims_All_Genomes_13V_1300_2000.txt', sep=',',header=None)
#Eliminar los ultimos 40 elementos del documento 4
doc_4 = doc_4.drop(range(700,740))

#Cargar quinto archivo
doc_5 = pd.read_csv('sims_All_Genomes_13V_2000_2350.txt', sep=',',header=None)
#Eliminar los ultimos 45 elementos del documento 5
doc_5 = doc_5.drop(range(350,395))

#Cargar sexto archivo
doc_6 = pd.read_csv('sims_All_Genomes_13V_2350_3222.txt', sep=',',header=None)

#Concatenar todos los archivos
datos = pd.concat([doc_1,doc_2,doc_3,doc_4,doc_5,doc_6])

#Modificar valores nulos
datos = datos.fillna(0)

#Convertir data frame a un array
array_datos = datos.to_numpy()
```

```python
#Obtener matriz espejo

#Definir un contador que ayuda a definir cuanto se corre cada elemento
contador=0
#Definir el array que guardara la matriz espejo
m_Espejo=[]

#Extraer cada fila del df
for i in array_datos:
    # Convertir elementos en un lista
    lista = i.tolist()
    lista.insert(0,0)
    #Generar for para ir moviendo la fila
    for x in range(contador):
        #Quitar ceros del final y agregarlos al inicio
        lista.pop()
        #Insetar un 0 para generar columna extra
        lista.insert(0,0)
    #Aumentar en 1 el contador
    contador+=1
    #Agregar la fila modificada al array
    m_Espejo.append(lista)

#Agregar una ultima fila a la matriz para convertirla en una de 3222 * 3222
m_Espejo.append([0]*3222)

#Generar matriz transpuesta
m_Transpuesta = np.transpose(m_Espejo)

#Sumar todas las matrices para general la matriz final
m_Final = m_Espejo + m_Transpuesta

#Generar dataframe y exportarlo
dataframe = pd.DataFrame(m_Final)
dataframe.to_csv(r'D:\Elementos Programacion\Cursos\Machine Learning\Ejercicios_Curso\Tareas_Data_Science_y_Machine_Learning\Kmeans\Datos_Reformados.txt',index=False)
```
